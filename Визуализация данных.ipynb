# Exploratory Data Analysis (EDA)
Предоставляем визуализацию данных

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('laptop.csv')
df
```

```python
plt.scatter(df['price'], df['brand'],c='slategrey')
plt.xlabel("Цена")
plt.ylabel("Фирма")
```

Такая диаграмма знакомит нас с распределением цен на ноутбуки в зависимости от их производителя, и Lenovo как будто бы занимает лидирующую позицию по числу ноутбуков, а Rikor и Durabook, но это мы проверим в процессе последовательного анализа датасета.

```python
plt.scatter(df['price'], df['screen_size'],c='slateblue')
plt.xlabel("Цена")
plt.ylabel("Диагональ дисплея")
```

```python
x = np.array(df['price'])
y = np.array(df['operating_system'])

plt.scatter(x, y)
plt.show()
```

Получившаяся диаграмма рассеяния показывает, что наиболее разнообразный спектр цен присущ операционной системе Windows 11 Pro, у которой ценовая категория начинается со значения менее 50000 и растягивается более, чем до 300000. Здесть стоит учитывать, что данная операционная система используется в компьютерах разных брендов. При этом система Mac OS, которая используется только компанией Apple, стоИт на ноутбуках стоимостью 100000 рублей и более. Отсюда можно сделать вывод, что операционная система Mac OS дороже операционной системы Windows 11 Pro, тк включается в стоимость ноутбуков Apple, а вторая упомянутая стоит дешевле, и есть как на относительно недорогих ноутбуках (<50000 рублей), так и на относительно дорогих (300000-400000 рублей), но такая высокая цена, очевидно, определяется другими уникальными дорогими составляющими техники.

```python
grid = sns.jointplot(df['screen_size'], df['weight'], kind='reg')
grid.fig.set_figwidth(4)
grid.fig.set_figheight(4)

plt.show()
```

Очевидно, что между диагональю дисплея в дюймах и весом ноутбука существует положительная линейная связь, соответственно, 
чем больше диагональ дисплея ноутбука, тем больше его вес.

```python
counts = df['brand'].value_counts()
    
plt.figure(figsize=(26, 4))    
plt.title('brand')
sns.barplot(counts.index, counts.values)
    
plt.show()
```

Столбчатая диаграмма наглядно демонстирует, что наибольшее количество ноутбуков в датасете производит Lenovo, как и предполагалось, при этом в данных встречаются фирмы, число ноутбуков от которых равно 1, и это те самые Rikor и Durabook. 

```python
sns.pairplot(df)
```

На графиках пар видно, что связь между парами количественных переменных есть. Графики наглядно показывают степень зависимости между переменными, например, умеренную зависимость между ценой и ядрами процессора. 

```python
plt.figure(figsize=(12,10))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds,fmt='.2f')
plt.show()
```

Получившаяся так называемая тепловая карта подтверждает зависимости, отраженные в парных графиках выше, и позволяет сказать, что наиболее сильно коррелируют цена и количеством ядер процессора в штуках, цена и оперативная память в гигабайтах, оперативная память в гигабайтах и количество ядер процессора в штуках, и снова подтвержается заметная взаимосвязь между весом и диагональю дисплея в дюймах.

```python
grid = sns.jointplot(df['price'], df['screen_size'], kind='hex', color='orange')
grid.fig.set_figwidth(10)
grid.fig.set_figheight(10)

plt.show()
```

В данном графике можно увидеть, что самыми популярными ноутбуками являются ноутбуки с диагональю 15.6. Так же в основном цена этих ноутбуков находится на интервале от 50000 до 100000 рублей
