Таким же образом мы хотели получить значения размеров дисплея по ноутбукам брэнда Apple путем перебора значений в переменной sizes с помощью цикла for и каждое значение вывести на печать с помощью функции print(). 


```python
from scipy.stats import norm

# Выборки с диагоналями дисплеев ноутбуков указанных выше брендов
weights1 = []
weights2 = []

# Вычисляем средние значения выборок
mean1 = np.mean(weights1)
mean2 = np.mean(weights2)

# Вычисляем стандартные отклонения выборок
std1 = np.std(weights1, ddof=1)
std2 = np.std(weights2, ddof=1)

# Вычисляем статистику z
z = (mean1 - mean2) / np.sqrt((std1**2)/len(weights1) + (std2**2)/len(weights2))

# Вычисляем критическое значение и p-value
alpha = 0.05 # Уровень значимости
z_crit = norm.ppf(1 - alpha/2) # Критическое значение z-статистики
p_val = 2 * (1 - norm.cdf(abs(z))) # Двухсторонний p-value

# Выводим результаты
print("Z-statistic:", z)
print("Critical value of Z:", z_crit)
print("p-value:", p_val)

if abs(z) > z_crit or p_val < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не удалось отвергнуть нулевую гипотезу")
```

В данном коде мы вычисляем средние значения и стандартные отклонения для каждой выборки, затем вычисляем статистику z и сравниваем ее со значением критической точки и уровнем значимости. Если бы в нашем тесте значение статистики z превышало критическое значение или p-value был бы меньше уровня значимости, мы бы отвергли нулевую гипотезу о равенстве средних размеров ноутбуков двух фирм.
