 Причина по которой мы выбрали именно эти столбцы заключается в том, что в них представлена самая важная информация по представленным моделям ноутбуков и эти столбцы имеют наименьшее количество пропусков. 
 
 Например до выделения этих столбцов для дальнейшей работы, можно заметить такие столбцы как: покрытие крышки дисплея, цвет. Эти столбцы не дают никакой информации для нашей дальнейшей работы, а только награмождают таблицу.
 
 Далее в этой тетрадке мы обработали данные в оставшихся столбцах, привели их к соответствующему типу и заполнили пропуски.
 - Например, значения цен из obj привели к int. 
 - Так же мы убрали лишнюю информацию из ячеек. Например, в столбце name перед каждым названием модели стоит слово 'Ноутбук', хотя и так понятно, с какими данными мы работаем

```python
#Тут мы просто переписали названия столбцов, ниже все названия столбцов будут описаны подробно
new_df.rename(columns = {'Диагональ дисплея ':'screen_size', 'Разрешение дисплея ':'display_resolution', 'Процессор ':'cpu', 'Количество ядер процессора ': 'proccessor_cores', 'Серия процессора. ': 'set_of_cpu', 'Частота процессора ': 'cpu_frequency', 'Видеокарта ':'gpu', 'Оперативная память ':'ram', 'Тип носителя данных ': 'memory', 'Установленная ОС ':'operating_system', 'Вес ':"weight", 'Бренд':'brand'}, inplace = True )
```

## Описание столбцов
 - name - название модели
 - price - цена в рублях
 - screen_size - диагональ дисплея в дюймах
 - display_resolution - разрешение дисплея
 - cpu - Процессор
 - proccessor_cores - количество ядер процессора в штуках
 - set_of_cpu - серия процессора
 - cpu_frequency - частота процессора в ГГц
 - gpu - видеокарта
 - ram - оперативная память в гигабайтах
 - memory - тип носителя данных - тип диска(ssd/hdd)
 - operating_system - установленная ОС - программное обеспечение ноутбука
 - weight - масса ноутбука; до обработки таблицы были значения, выраженные как в граммах, так и килограммах. В итоговом варианте в граммах
 - brand - компания-производитель

```python
# Ниже мы меняли тип данных со obj в тех столбцах, где это было необходимо
pd.options.mode.chained_assignment = None
new_df['name'] = new_df['name'].str.replace('Ноутбук ','')
```

```python
new_df['price'] = new_df['price'].str.replace(' руб.','')
# Это просто предупреждение от питона, а не ошибка :)
```

```python
new_df = new_df.dropna(subset=['price'])
# Так же мы решили выкинуть несколько ноутбуков, о которых не было информации о цене
new_df['price'] = new_df['price'].astype(int)
```

```python
new_df['screen_size'] = new_df['screen_size'].str.split('дюйм', n=1).str[0]
new_df['screen_size']=pd.to_numeric(new_df['screen_size'])
```

```python
new_df['proccessor_cores'] = new_df['proccessor_cores'].str.split('/', n=1).str[0]
new_df['proccessor_cores'] = pd.to_numeric(new_df['proccessor_cores'], downcast = 'integer')
```

```python
new_df['cpu_frequency'] = new_df['cpu_frequency'].str.split(' ГГц').str[0]
new_df['cpu_frequency'] = new_df['cpu_frequency'].str.split('+').str[0]
new_df['cpu_frequency'] = pd.to_numeric(new_df['cpu_frequency'], downcast = 'integer')
```

```python
new_df['ram'] = new_df['ram'].str.split(' Gb').str[0]
new_df['ram'] = pd.to_numeric(new_df['ram'], downcast = 'integer')
```

```python
# С помощью цикла все ячейки перевели в единую систему - граммы, убрали лишний текст и преобразовали тип данных в ячейках
def convert_to_grams(value):
    if pd.isna(value):
        return value
    else:
        unit = value.split()[-1]
        number = float(value.replace(',','.').split()[0])
        if unit == 'кг':
            number *= 1000
        return number
new_df['weight']=new_df['weight'].apply(convert_to_grams)
```

## Заполнение пропусков
 Из сводки можно заметить, что пропущенных значений достаточно мало, это хорошо, но надо бы дозаполнять таблицу.
 
 Пропуски во всех столбцах, в которых информация содержится в виде object, мы заменили на первую моду этого столбца, так как ни среднюю, ни медиану применить к типу object нельзя, так как они обычно применяются к числовым данным.
 Однако, пропуски в столбцах screen_size, proccessor_cores, cpu_frequency и ram мы тоже заменили на первую моду, так как далеко не все значения, которые могут получиться, например, с помощью средней реально существуют. Так как у этих характеристик ноутбуков обычно определенный перечень значений.
 
 А вот пропуски в массах ноутбуков мы заменили на среднее значение в столбце, так как масса может быть абсолютно любой

```python
new_df['screen_size'] = new_df['screen_size'].fillna(new_df['screen_size'].mode()[0])
```

```python
new_df['display_resolution'] = new_df['display_resolution'].fillna(new_df['display_resolution'].mode()[0])
```

```python
new_df['cpu'] = new_df['cpu'].fillna(new_df['cpu'].mode()[0]) 
```

```python
new_df['proccessor_cores'] = new_df['proccessor_cores'].fillna(new_df['proccessor_cores'].mode()[0])
```

```python
new_df['set_of_cpu'] = new_df['set_of_cpu'].fillna(new_df['set_of_cpu'].mode()[0]) 
```

```python
new_df['cpu_frequency'] = new_df['cpu_frequency'].fillna(new_df['cpu_frequency'].mode()[0]) 
```

```python
new_df['gpu'] = new_df['gpu'].fillna(new_df['gpu'].mode()[0]) 
```

```python
new_df['ram'] = new_df['ram'].fillna(new_df['ram'].mode()[0]) 
```

```python
new_df['memory'] = new_df['memory'].fillna(new_df['memory'].mode()[0]) 
```

```python
new_df['operating_system'] = new_df['operating_system'].fillna(new_df['operating_system'].mode()[0]) 
```

```python
new_df['weight'] = new_df['weight'].fillna(new_df['weight'].mean())
```
